6.Implement a Python program to fit a Simple Linear Regression model to
the following sample data:
  ● X = [1, 2, 3, 4, 5]
  ● Y = [2.1, 4.3, 6.1, 7.9, 10.2]
Plot the regression line over the data points.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5])
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# Reshape X to be a 2D array for scikit-learn
X_reshaped = X.reshape(-1, 1)

# Create a Linear Regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X_reshaped, Y)

# Get the slope (coefficient) and intercept of the regression line
slope = model.coef_[0]
intercept = model.intercept_

print(f"Regression Equation: Y = {slope:.2f}X + {intercept:.2f}")

# Predict Y values using the fitted model
Y_pred = model.predict(X_reshaped)

# Plot the original data points and the regression line
plt.figure(figsize=(8, 6))
plt.scatter(X, Y, color='blue', label='Original Data Points')
plt.plot(X, Y_pred, color='red', label='Regression Line')
plt.title('Simple Linear Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```

### Residuals Plot for Heteroscedasticity Assessment

We will train a simple linear regression model on the new data, calculate the residuals, and then create a residuals plot. This plot helps us visually inspect for heteroscedasticity.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# New sample data
X_new = np.array([10, 20, 30, 40, 50])
Y_new = np.array([15, 35, 40, 50, 65])

# Reshape X to be a 2D array for scikit-learn
X_new_reshaped = X_new.reshape(-1, 1)

# Create a Linear Regression model
model_residuals = LinearRegression()

# Fit the model to the data
model_residuals.fit(X_new_reshaped, Y_new)

# Predict Y values using the fitted model
Y_new_pred = model_residuals.predict(X_new_reshaped)

# Calculate residuals
residuals = Y_new - Y_new_pred

# Create the residuals plot
plt.figure(figsize=(10, 6))
plt.scatter(Y_new_pred, residuals, color='purple', alpha=0.7)
plt.axhline(y=0, color='red', linestyle='--', linewidth=2, label='Zero Residual Line')
plt.title('Residuals Plot')
plt.xlabel(r'Predicted Values ($\hat{Y}$)') # Using raw string
plt.ylabel(r'Residuals (Y - $\hat{Y}$)') # Using raw string
plt.grid(True)
plt.legend()
plt.show()
```

#### Assessing Heteroscedasticity from the Residuals Plot

To assess heteroscedasticity (non-constant variance of residuals) from the residuals plot:

1.  **Look for patterns:** If the spread of the residuals (the vertical distance of the points from the horizontal zero line) changes systematically as the predicted values (on the x-axis) increase, it suggests heteroscedasticity.
    *   **Heteroscedasticity indicators:** A 'cone' shape (residuals widening or narrowing), a fan shape, or any clear pattern in the spread of points indicate heteroscedasticity.
    *   **Homoscedasticity indicators:** If the residuals are randomly scattered around the zero line with a roughly constant spread across all predicted values, it indicates homoscedasticity (constant variance), which is an assumption of linear regression.

In the plot above, observe if the purple points are evenly distributed around the red dashed line, or if their vertical spread changes as you move from left to right.

### Addressing Heteroscedasticity in House Price Prediction

Heteroscedasticity occurs when the variance of the error terms (residuals) is not constant across all levels of the independent variables. In real estate, this might mean that prediction errors are larger for more expensive homes or for homes in certain locations.

**1. Detection:**
*   **Residuals Plot:** The primary visual tool. I'd plot the residuals against predicted values (or independent variables). A 'funnel' or 'cone' shape indicates heteroscedasticity.
*   **Statistical Tests:** For more rigorous detection, I'd use tests like the **Breusch-Pagan test** or **White test** to statistically confirm the presence of heteroscedasticity.

**2. Addressing Heteroscedasticity:**
*   **Data Transformations:**
    *   **Log Transformation:** Applying a log transformation to the dependent variable (house price) or highly skewed independent variables can often stabilize the variance. For example, `log(Price)` instead of `Price`.
    *   **Square Root Transformation:** Another option for positive, skewed data.
*   **Weighted Least Squares (WLS):** If the pattern of heteroscedasticity is known or can be estimated, WLS assigns different weights to observations, giving less weight to observations with larger variances. This is more complex but can be very effective.
*   **Robust Standard Errors:** Also known as Huber-White standard errors. These don't eliminate heteroscedasticity but adjust the standard errors of the coefficients, making hypothesis tests and confidence intervals more reliable even in its presence. This is often a pragmatic first step if transformations don't fully resolve the issue or are not desired.

### Addressing Multicollinearity in House Price Prediction

Multicollinearity occurs when two or more independent variables in a regression model are highly correlated with each other. In house price prediction, this could happen if 'Area' and 'Number of Rooms' are strongly correlated (larger houses tend to have more rooms) or if 'Location' features (e.g., proximity to school, crime rate) are intertwined.

**1. Detection:**
*   **Correlation Matrix:** A simple way to identify highly correlated pairs of independent variables.
*   **Variance Inflation Factor (VIF):** This is the most common and robust method. VIF measures how much the variance of an estimated regression coefficient is inflated due to multicollinearity. A VIF value above 5 or 10 typically indicates significant multicollinearity.

**2. Addressing Multicollinearity:**
*   **Feature Selection:**
    *   **Remove Highly Correlated Variables:** If two variables are highly correlated, I'd choose one to keep based on domain knowledge, predictive power, or simplicity. For example, if 'total living area' and 'number of bedrooms' are highly correlated, I might keep 'total living area' as it often has a more direct impact on price.
    *   **Stepwise Regression:** Automated methods (forward, backward, bidirectional) can help select a subset of variables, though care must be taken to avoid overfitting.
*   **Dimensionality Reduction:**
    *   **Principal Component Analysis (PCA):** This technique transforms the correlated variables into a smaller set of uncorrelated components (principal components) that capture most of the original variance. The new components can then be used in the regression model.
*   **Regularization Techniques:**
    *   **Lasso (L1 Regularization):** This technique adds a penalty to the regression model proportional to the absolute size of the coefficients. It can shrink some coefficients to exactly zero, effectively performing feature selection and helping to mitigate multicollinearity.
    *   **Ridge (L2 Regularization):** This technique adds a penalty proportional to the square of the magnitude of coefficients. It shrinks correlated predictor coefficients towards each other, stabilizing them and reducing their variance, without necessarily setting them to zero.
*   **Collect More Data:** In some cases, increasing the sample size can help reduce the impact of multicollinearity, as it can provide more independent variation across the correlated features.

By systematically addressing both heteroscedasticity and multicollinearity, I can build a more reliable and interpretable model for predicting house prices, leading to more accurate insights and business decisions for the real estate company.
